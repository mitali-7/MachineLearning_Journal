```python
# Operations like Union, Intersection, Diff

set1 = {1, 3, 4, 5, 7, 8}
set2 = {2, 4, 6, 8}

print(set1.union(set2))

print(set1.intersection(set2))

set1.intersection_update(set2)

print(set1)

set3 = {3, 4, 5, 7}
set4 = {5, 7, 8, 9}

# diff - elts present in set1 but arent present in set2
print(set3.difference(set4))

# symmetric diff - unique elts from both sets are kept
print(set3.symmetric_difference(set4))
```

```python
# Nested Dict

student = {
    "stud1" : {"name" : "Mitali", "id" : "25MCA0126"},
    "stud2" : {"name": "Adi", "id": "25MCA0127"}
}

print(student)

print(student["stud2"]["id"])

print("\n")

# iterating thru nested dict
for num, details in student.items():
    print(num)
    for key, val in details.items():
        print(f"{key} : {val}")        
```

```python
# Count frequency of elts in a list

lst = ['a', 'a', 'a', 'b', 'b', 'f', 'f', 'g']
freq = {}

for num in lst:
    if num in freq:
        freq[num]+=1
    else:
        freq[num] = 1  

for k, v in freq.items():
    print(f"{k} : {v}")
```

```python
# Password strength checker

def chk_pswd(password):
    if len(password) < 8:
        return False
    if not any (char.isdigit() for char in password ):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in '!@#$%^&*()_-' for char in password):
        return False
    return True
# Can also return None to display nothing


chk_pswd("Abcd@123")
```

```python
# Totaling price of items in cart

def price_total(cart):
    total = 0
    for items in cart:
        total += items['price'] * items['quantity']
    return total    

cart = [{'name': "Apple", 'price': 5, 'quantity': 3},
        {'name': "Ball", 'price': 50, 'quantity': 5},
        {'name': "Cap", 'price': 300, 'quantity': 2}]    

print(price_total(cart))
```

```python
# Check string is palindrome

s = "Hello world dlrow olleH"

def chk_palin(s):
    s = s.lower().replace(" ", "")
    return s == s[::-1]

print(chk_palin(s))
```

```python
class dog:
    # default age is 1
    def __init__(self, name, age = 1):
        self.name = name
        self.age = age

    def bark(self):
        print(f"{self.name} is barking")   

dog1 = dog('Lucy', 2)
dog2 = dog('Bobby')

print(dog2.age)
dog1.bark()
dog2.bark()
```

```python
# designing a bank account in OOPs

class account:
    def __init__(self, owner, bal = 0):
        self.owner = owner
        self.bal = bal


    def deposit(self, amt):
        self.bal+=amt 
        print(f"{amt} is deposited. New balance is {self.bal}")


    def withdraw(self, amt):
        if amt>self.bal:
            print("Low balance")
        else:
            self.bal -= amt
            print(f"{amt} is withdrawn. New balance is {self.bal}")  

    def chkbal(self):
        print(f"Current balance is {self.bal}")


cust1 = account('Mitali', 500)
cust1.chkbal()
cust1.deposit(200)
cust1.withdraw(100)   
     
```
