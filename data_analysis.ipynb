```python
#  Numpy Vectorized Operations
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([10, 20, 30, 40, 50])

print("Addition: ", arr1+arr2)
print("Subtraction", arr1 - arr2)
print("Multiply", arr1*arr2)
print("Division", arr2//arr1)
```

```python
# checking missing values - no true values means no missing val
df.isnull()
```

```python
# checking col wise if any null values are encountered in it
df.isnull().any()
```

```python
# gives count of null vals in each col
df.isnull().sum()
```

```python
# filling null vals with 0
df.fillna(0)
```

```python
# fill null val with mean val of that specifc col
age_na = df['age'].fillna(df['age'].mean())
age_na
```

```python
# renaming cols

df = df.rename(columns={"age" : "AGE"})
df.head()
```

```python
df.dtypes
```

```python
# changing datatype of col - Age from float to int

df['int_age'] = df['AGE'].fillna(df['AGE'].mean()).astype(int)
df

```

```python
# Aggregation - group by

grped_mean = df.groupby('department')['marks'].mean()
grped_mean
```

```python
# grouping by 2 attributes

grp_sum = df.groupby(['AGE', 'department'])['marks'].mean()
print(grp_sum)
```

```python
# aggregate multiple functions

grp_multiple = df.groupby('department')['marks'].agg(['mean', 'sum', 'count'])
print(grp_multiple)
```

```python
# MAtplotlib
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# creating a line plot
plt.plot(x, y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line graph')
```

```python
# formatting the graph
plt.plot(x, y, color='red',linestyle='--', marker='*')
```

```python
#  bar plot

category = ['A', 'B', 'C', 'D']
values = [4, 3, 6, 2]

plt.bar(category, values, color='purple')
plt.title('BAr Plot')
```
